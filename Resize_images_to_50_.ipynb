```python
import os
import cv2
from google.colab.patches import cv2_imshow
```

```python
# Folder path containing the images
folder_path = "/content/images"
output_path = "/content/resize_images"
```

```python
# Get a list of all the image files in the folder
image_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
```

```python
# Iterate over each image file
for image_file in image_files:
    # Load the image
    image_path = os.path.join(folder_path, image_file)
    image = cv2.imread(image_path)

    half = cv2.resize(image, (0, 0), fx = 0.1, fy = 0.1)

    # Save the grayscale image
    grayscale_image_path = os.path.join(output_path, f"bw_{image_file}")
    cv2.imwrite(grayscale_image_path, half)

    # Display the converted image

    cv2_imshow(half)
    cv2.waitKey(0)

# Close all windows
cv2.destroyAllWindows()
```

```python
import shutil
import os.path

# Creating the ZIP file
archived = shutil.make_archive('/content/resized_image', 'zip', '/content/resize_images')

if os.path.exists('/content/resized_image.zip'):
   print(archived)
else:
   print("ZIP file not created")
```
